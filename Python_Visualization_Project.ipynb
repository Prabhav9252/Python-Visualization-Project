```python
# lets move for the salary distribution analysis
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Prabhav9252/Python-Visualization-Project/refs/heads/main/df.csv")
```

```python
df.drop(columns="Unnamed: 0",inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['salary'],
             bins = 20,
             color = 'teal',
             edgecolor='black',
             alpha=0.7,
             kde=True
            )
plt.title('Salary Distribution across Employees',fontsize = 14,pad=10)
plt.xlabel('Salary',fontsize=12)
plt.ylabel('No of employees', fontsize=12)
plt.grid(True,alpha=0.3)
plt.show()
```

# 2 Salary Comparison Based on Promotion Status

```python
melted_data = pd.melt(df,
        id_vars = ['promotion_hr_based'],
        value_vars=['salary','final_incremented_salary'],
        var_name = 'salary_type',
        value_name = 'Salary'
       )
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='promotion_hr_based',y='Salary',hue='salary_type',palette='Set2',data=melted_data)
plt.title('Salary Comparision based on Promotion Status', fontsize = 14)
plt.xlabel('Promotion HR Data', fontsize=12)
plt.ylabel('Salary',fontsize=12)
plt.legend(title='Salary Type')
plt.show()
```

## 3. Increment Percentage Evaluation

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='promotion_hr_based',y='increment_percentage',
           hue='promotion_hr_based', palette ='bright',data=df)
plt.title('Increment Percentage', fontsize = 14)
plt.xlabel('Promotion HR Based',fontsize=12)
plt.ylabel('Increment Percentage (%)', fontsize=12)
plt.show()
```

## 4. Relationship between Experience and Salary Growth

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='experience_years', y='salary',data=df,hue='promotion_hr_based',size='increment_percentage',style='promotion_hr_based',sizes=(50,100))
plt.title("Experience Vs Salary by Promotion Cycle", fontsize = 14)
plt.xlabel("Years of Experience", fontsize = 12)
plt.ylabel("Salary",fontsize=12)
plt.legend(title = 'Promotion Status',bbox_to_anchor=(1.05,1),loc='upper left')
```

## Sales vs Productivity

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='time_productive',y='sales',hue='experience_years',data=df,sizes=(50,100),palette='viridis',
               size='experience_years')
plt.title('Sales Vs Productivity', fontsize = 14,pad=10)
plt.xlabel('Productivity',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend(title='Years of Experience',bbox_to_anchor=(1.05,1),loc = 'upper left')
plt.tight_layout()
```

## Average sales and productivity by Experience

```python
grouped_data = df.groupby('experience_years').mean(numeric_only=True)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='sales',y='time_productive',data=grouped_data,hue='experience_years',palette='deep')
plt.legend(title='Years of Experience',bbox_to_anchor=(1.05,1),loc='upper left')
```

## Top 5 Performers

```python
top_5 = df[['first_name','sales']].sort_values(by='sales',ascending=False).head(5)
top_5
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='first_name',y='sales',data=top_5,palette='Pastel1',hue='first_name',legend=False)
plt.title('Top 5 Performer',fontsize=14,pad = 10)
plt.xlabel('Emp Name', fontsize = 12)
plt.ylabel('Sales',fontsize = 12)
plt.xticks(rotation=45)
plt.show()
```

So Yuki has multiple sales values, and Seaborn tries to calculate:

## Experience Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df['experience_years'],
            bins=20,color='Teal',
            edgecolor='black',
            alpha=0.7,
            kde=True
            )
plt.title('Employees Experience Distribution',fontsize=14,pad=10)
plt.xlabel('Years of Experience',fontsize=12)
plt.ylabel('No of Employees',fontsize=12)
plt.show()
```

## Promotion Hours and Experience

```python
promotion_data = df.groupby('experience_years')['time_productive'].mean().reset_index()
```

```python
promotion_data
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='experience_years',y='time_productive',data=promotion_data,marker='o',color='Orange')
plt.title('Average Promotion Hours by Experience',fontsize=14,pad=10)
plt.xlabel('Years of Experience',fontsize=12)
plt.ylabel('Average Promotion Hours',fontsize=12)
plt.show()
```

## Bubble Chart (Experience vs Years)

```python
plt.figure(figsize=(10,6))
fig1 = px.scatter(df,x='experience_years',y='sales',color='time_productive',hover_data=['first_name'],title='Bubble Chart (Experience vs Years)',color_continuous_scale=px.colors.sequential.Blues)
fig1.update_layout(xaxis_title='Years of Experience',yaxis_title='Sales',coloraxis_colorbar_title='Productivity Time(Hours)')
fig1.show()
plt.show()
```
